```python
#import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load bangalore house dataset
df = pd.read_csv("Bengaluru_House_Data.csv")
df.head()
```

Exploratory Data Analysis

```python
#drop unwanted columns
cols = ["area_type","availability","society","balcony"]
df = df.drop(cols,axis=1)
```

```python
#Drop duplicate rows
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#cleaning location column

#i) remove leading and trailing spaces
df["location"] = df["location"].str.strip()

#ii)handling nan values in location with top 2nd location-->Sarjapur  Road
df["location"] = df["location"].fillna("Sarjapur  Road")

#iii)Grouping the locations with less than 10 houses into others category
loc = df["location"].value_counts()
loc_less_than_ten = loc[loc<=10]
df["location"] = df["location"].apply(lambda x : "others" if x in loc_less_than_ten else x)

```

```python
#cleaning size column

#i)handling nan values in size column
df["size"] = df["size"].fillna("2 BHK")

#ii) extract only the bhk number and remove unwanted chars
df["size"] = df["size"].apply(lambda x : int(str(x).split()[0]))
df["size"].unique()

#iii)renaming size to BHK
df.rename(columns={"size":"BHK"},inplace=True)
```

```python
#handling total_sqft column
def clean_total_sqtf(sqft):
    try:
        l = sqft.split("-")
        if len(l) == 2:
            return (float(l[0])+float(l[1])) / 2
        else:
            return float(l[0])
    except:
        return None
    
df["total_sqft"] = df["total_sqft"].apply(clean_total_sqtf)

#handling nan values in total_sqft

df["total_sqft"] = df["total_sqft"].fillna(round(df["total_sqft"].mean()))
```

```python
#handling nan values in bath
med = df["bath"].median()
df["bath"] = df["bath"].fillna(med)
df["bath"].unique()
```

Handling outliers

```python
#create a new column price_per_sqft
#formula : price_per_sqft = price/total_sqft
df["price_per_sqft"]=(df["price"]*100000) / df["total_sqft"]
```

```python
#handling outliers in total_sqft

# Observation : In total_sqft column we have houses less than 300 sqft those are outliers
# In real time houses starts with minimum of 300 sqft

# formula : total_sqft/bhk >= 300 --> it keeps all the realistic total sqft 
df = df[df["total_sqft"]/df["BHK"] >= 300]
```

```python
df.describe()
```

All unrealistic rows with total_sqft less than 300 got removed

```python
#handling outliers in BHK
#sns.boxplot(x="BHK",data=df)
df = df[df["BHK"] <= 6]
sns.boxplot(x="BHK",data=df)
```

```python
#handling outliers in bath column
df[df["bath"]>df["BHK"]][["BHK","bath"]]
# Observation: Here no.of bath count is suspiciously high compared to bhk number these are outliers
df = df[df["bath"]<df["BHK"]+2]  #removes (outliers) unrealistic values where no.of bath count is suspiciously high compared to bhk
```

```python
#Handling outliers in price_per_sqft
#sns.boxplot(x="price_per_sqft",data=df)
#IQR Method
q1 = df["price_per_sqft"].quantile(0.25)
q3 = df["price_per_sqft"].quantile(0.75)
IQR = q3-q1
lower = q1-0.5*IQR
upper = q3+0.5*IQR
df = df[(df["price_per_sqft"]>=lower) & (df["price_per_sqft"]<=upper)]
sns.boxplot(x="price_per_sqft",data=df)
```

```python
#drop price_per_sqft column
df = df.drop("price_per_sqft",axis=1)
```

```python
#reset index
df.reset_index(inplace=True)
df=df.drop("index",axis=1)
df
```

```python
plt.scatter(df["total_sqft"],df["price"])
plt.xlabel("Total sqft")
plt.ylabel("Price")
plt.title("Total sqft vs price")
```

```python
data = df.iloc[:,1:].corr()
sns.heatmap(data,annot=True)
```

```python
#average price by location
data=df.groupby("location")["price"].mean().sort_values()
data

```

Model Training

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
#Encoding location column
encoder = LabelEncoder()
df["encoded_location"] = encoder.fit_transform(df["location"])
```

```python
#Divide data into input and output features
X = df.drop(["location","price"],axis=1)
y = df["price"]
```

```python
#split the data for training and testing
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size=0.3,random_state=42)
```

Linear Regression

```python
#model training
LR = LinearRegression()
LR.fit(Xtrain,ytrain)
```

```python
#Predictions
ypred = LR.predict(Xtest)
ypred
```

```python
print("Training Eff: ",LR.score(Xtrain,ytrain))
print("Training Eff: ",LR.score(Xtest,ytest))
print("R2 score: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

Decision Tree

```python
DT = DecisionTreeRegressor()
DT.fit(Xtrain,ytrain)
```

```python
#Predictions
ypred = DT.predict(Xtest)
ypred
```

```python
print("Training Eff: ",DT.score(Xtrain,ytrain))
print("Training Eff: ",DT.score(Xtest,ytest))
print("R2 score: ",r2_score(ytest,ypred))
print("MAE: ",mean_absolute_error(ytest,ypred))
```

Decision Tree is overfitting cause the data is linear

here linear regression working fine than decision tree
